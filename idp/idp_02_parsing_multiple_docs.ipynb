# IDP Lesson 02: Parsing Multiple Documents

```sql
CREATE OR REPLACE TEMP VIEW raw_structured_doc AS
SELECT 
path,
content 
FROM read_files('/Volumes/idp/default/multiple_docs_parsing')
```

```sql
CREATE OR REPLACE TEMP VIEW parsed_structured_doc AS
SELECT 
path,
ai_parse_document(content) AS parsed_content
FROM raw_structured_doc
```

```sql
SELECT * FROM parsed_structured_doc
```

```sql
SELECT *,
ai_classify(cast(parsed_content as string), 
array('Medical Claim', 'Medical Invoice', 'Lab Report', 'Clinical Note', 'Administrative Document')) AS label
FROM parsed_structured_doc
```
