```sql
CREATE OR REPLACE TABLE parsed_data AS
SELECT 
path,
ai_parse_document(content) as parsed_content
FROM read_files('/Volumes/idp/default/idp_final_project')
```

```sql
SELECT * FROM parsed_data
```

```sql
CREATE OR REPLACE TABLE pretty_data AS
SELECT 
    path,
    concat_ws('\n',
        transform(
            try_cast(parsed_content:document:elements AS ARRAY<VARIANT>), 
            e -> coalesce(try_cast(e:content AS STRING), '')
        )
    ) AS doc_text
FROM parsed_data;
```

```sql
CREATE OR REPLACE TABLE classified_data AS
SELECT 
*,
ai_classify(doc_text, array('Invoice', 'Purchase Order', 'Receipt', 'Other')) AS doc_classification
FROM pretty_data
```

**Invoice Extraction**

```sql
CREATE OR REPLACE TABLE invoice_data AS
SELECT *,
ai_extract(doc_text, 
ARRAY('Vendor_Name',
      'Invoice_Number',
      'Invoice_Date',
      'Due_Date',
      'Payment_Method',
      'Total')) AS extracted
FROM classified_data
WHERE doc_classification = 'Invoice';

SELECT * FROM invoice_data;
```

**Purchase Order Extraction**

```sql
CREATE OR REPLACE TABLE po_data AS
SELECT *,
ai_extract(doc_text, 
ARRAY('Merchant_Name',
      'PO_Number',
      'PO_Date',
      'Total')) AS extracted
FROM classified_data
WHERE doc_classification = 'Purchase Order';

SELECT * FROM po_data;
```

**Receipt Extraction**

```sql
CREATE OR REPLACE TABLE receipt_data AS
SELECT *,
ai_extract(doc_text, 
ARRAY('Merchant_Name',
      'Receipt_Number',
      'Transaction_Date',
      'Total')) AS extracted
FROM classified_data
WHERE doc_classification = 'Receipt';

SELECT * FROM receipt_data;
```

**Create Schema To Store Extracted Data**

```sql
CREATE SCHEMA IF NOT EXISTS idp.finance
```

**Append `invoices`**

```sql
CREATE OR REPLACE TABLE idp.finance.invoices AS
SELECT path,
extracted.Vendor_Name AS Vendor,
extracted.Invoice_Number AS Invoice_Number,
extracted.Invoice_Date AS Invoice_Date,
extracted.Due_Date AS Due_Date,
extracted.Payment_Method AS Payment_Method,
extracted.Total AS Total
FROM invoice_data
```

```sql
SELECT * FROM idp.finance.invoices
```

**Append `purchase_order`**

```sql
CREATE OR REPLACE TABLE idp.finance.purchase_order AS
SELECT path,
extracted.Merchant_Name AS Merchant_Name,
extracted.PO_Number AS PO_Number,
extracted.PO_Date AS PO_Date,
extracted.Total AS Total
FROM po_data
```

```sql
SELECT * FROM idp.finance.purchase_order
```

**Append `receipt`**

```sql
CREATE OR REPLACE TABLE idp.finance.receipt AS
SELECT path,
extracted.Merchant_Name AS Merchant_Name,
extracted.Receipt_Number AS Receipt_Number,
extracted.Transaction_Date AS Transaction_Date,
extracted.Total AS Total
FROM receipt_data
```

```sql
SELECT * FROM idp.finance.receipt
```
